```python
from sqlalchemy import create_engine
import pandas as pd
engine = create_engine('postgresql://nick@localhost:5432/mlb_stats')
```

```python
df = pd.read_sql_table('atbat', engine)
df2 = pd.read_sql_table('player', engine)
pitchers = df.merge(df2, left_on="pitcher", right_on="id")
batters = df.merge(df2, left_on="batter", right_on="id")
```

```python
batters['batter_full_name'] = batters['first_name'] + ' ' + batters['last_name']
pitchers['pitcher_full_name'] = pitchers['first_name'] + ' ' + pitchers['last_name']
at_bat_counts = batters['batter_full_name'].value_counts()
faced_batters = pitchers['pitcher_full_name'].value_counts()
len(at_bat_counts[at_bat_counts > 100])
```

```python
len(faced_batters[faced_batters > 100])
```
