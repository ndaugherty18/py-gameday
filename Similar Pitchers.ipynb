```python
from sqlalchemy import create_engine
import pandas as pd
engine = create_engine('postgresql://nick@localhost:5432/mlb_stats')
```

```python
sql = "select pitcher, count(pitcher) from pitch group by pitcher having count(pitcher) > 999"
pitches = pd.read_sql_query(sql, engine)
ids = pitches['pitcher'].tolist()
```

```python
ids_str = ','.join([str(id) for id in ids])
sql2 = "select * from pitch where pitcher in (%s)" % ids_str
df = pd.read_sql_query(sql2, engine)
df
```
